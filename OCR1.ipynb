```python
import os
import pandas as pd

# Load the vocabulary mapping
hindi_vocab_path = "/content/hindi/val.txt"

with open(hindi_vocab_path, "r") as f:
    hindi_vocab_lines = f.readlines()

hindi_vocab = pd.DataFrame([line.split() for line in hindi_vocab_lines], columns=["vocab_id", "word"])
hindi_vocab_dict = hindi_vocab.set_index("vocab_id")["word"].to_dict()

# Function to extract labels from image paths
def extract_labels(image_path, hindi_vocab_dict):
    # Extract the writer ID, page number, and image name from the image path
    writer_id, page_number, image_name = image_path.split("/")[-3:]
    page_number = int(page_number.split(".")[0])

    # Load the corresponding text file
    text_file_path = f"HindiSeg/{writer_id}/{page_number}.txt"
    with open(text_file_path, "r") as f:
        lines = f.readlines()

    # Find the line corresponding to the image number
    for line in lines:
        if line.startswith(f"{image_name}"):
            vocab_id = int(line.split()[1])
            label = hindi_vocab_dict[vocab_id]
            return label

    # If no match found, return None
    return None

# Function to load data from the val.txt file
def load_val_data(val_data_path):
    with open(val_data_path, "r") as f:
        lines = f.readlines()

    data = []
    for line in lines:
        image_path, label = line.strip().split()
        data.append((image_path, label))
    return data

# Load val data
val_data_path = "/content/hindi/val.txt"
val_data = load_val_data(val_data_path)

# Convert data to DataFrame
val_data = pd.DataFrame(val_data, columns=["image_path", "label"])

# Save the processed data
val_data.to_csv("/content/file/processed_val.csv", index=False)
```

```python
import os
import pandas as pd

# Example Hindi word list (replace this with the actual list you're using)
#hindi_words = ['शब्द1', 'शब्द2', 'शब्द3', ...]  # Replace with your actual hindi_words list

def process_folder(folder_path, hindi_words):
    """Process a single folder and return a dataframe with the full image path and corresponding label."""
    text_file = None
    images = []

    # Search for the text file and images in the folder
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if file.endswith('.txt'):
            text_file = file_path
        elif file.endswith('.jpg'):
            images.append(file_path)  # Save the full file path instead of just the file name

    # Sort images to ensure they are processed in the order '1.jpg', '2.jpg', etc.
    images.sort(key=lambda x: int(os.path.splitext(os.path.basename(x))[0]))  # Sort by the image number

    # If no text file is found, return None
    if not text_file:
        return None

    # Read the numbers from the text file
    with open(text_file, 'r') as f:
        labels = [int(line.strip()) for line in f.readlines()]

    # Create a dataframe mapping the full image path to Hindi words
    data = []
    for img_file, label in zip(images, labels):
        # Use the label as the index to retrieve the corresponding Hindi word (label - 1 for zero-based indexing)
        word = hindi_words[label - 1]  # Minus 1 because of zero-indexing
        data.append({'image_path': img_file, 'label': word})

    return pd.DataFrame(data)

def traverse_and_process(root_folder, hindi_words):
    """Traverse subfolders and process the data, concatenating results into a master dataframe."""
    master_df = pd.DataFrame()

    # Traverse the folders one level deep
    for subfolder in os.listdir(root_folder):
        subfolder_path = os.path.join(root_folder, subfolder)

        # If the path is a directory, process the folder
        if os.path.isdir(subfolder_path):
            val_test = process_folder(subfolder_path, hindi_words)

            if val_test is not None:
                master_df = pd.concat([master_df, val_test], ignore_index=True)

    return master_df
```
